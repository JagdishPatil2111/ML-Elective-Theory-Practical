```python
"""
Q1. Consider the attached dataset (Mult_reg.png) of power consumption containing 
(4 rows and 3 column). Define the function for computing the regression 
coefficients. Print out the coefficients.
"""
import numpy as np
x = np.array([[1,10,2],[1,5,42],[1,4.5,31],[1,6,35]])
y = np.array([[0.5],[2.5],[2.2],[1.8]])
def coefficient (x,y):
  beta = np.dot((np.dot((np.linalg.inv(np.dot((x.T),x))),x.T)),y)
  return beta
beta = coefficient(x,y)
print("B0 : ",beta[0],"\nB1 : ",beta[1],"\nB2 : ",beta[2])
```

```python
"""
Q2. Define the function for computing energy consumption for given regression 
coefficients as inputs as follows. Print the output describing energy 
required (KW) for given temp and no. of persons.
   I. temp = 7.5 and No. of persons in room = 131
  II. temp = 3.5 and No. of persons in room = 45
 III. temp = 7.2 and No. of persons in room = 27
"""
x = np.array([[1,10,2],[1,5,42],[1,4.5,31],[1,6,35]])
y = np.array([[0.5],[2.5],[2.2],[1.8]])
def coefficient (x,y):
  beta = np.dot((np.dot((np.linalg.inv(np.dot((x.T),x))),x.T)),y)
  return beta
beta = coefficient(x,y)
x1 = np.array([1,7.5,131])
power1 = np.dot(x1,beta)

x2 = np.array([1,3.5,45])
power2 = np.dot (x2,beta)

x3 = np.array([1,7.2,27])
power3 = np.dot(x3,beta)

print("If outside temp is",x1[1]," and number of people in room is",int(x1[2]),
      "then energy consumption is ",power1[0],"kw")

print("If outside temp is",x2[1]," and number of people in room is",int(x2[2]),
      "then energy consumption is ",power2[0],"kW")

print("If outside temp is",x3[1]," and number of people in room is",int(x3[2]),
      "then energy consumption is ",power3[0], "kw")
```

```python
"""
Q3. Plot the 3D graph for power consumption data using multivariate regression 
coefficients(fit the hyperplane).
"""
import matplotlib.pyplot as plt
a = np.arange(0,10,0.5)
b = np.arange(0,100,1)
X11,X12 = np.meshgrid(a,b)
y_new = beta[0] + beta[1]*X11 + beta[2]*X12
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot_surface(X11, X12, y_new, cmap = 'Greens')
ax.set_xlabel("Outside temp")
ax.set_ylabel("No of people in room")
ax.set_zlabel("Power (ks)")
ax.set_title("Multilinear Regression")
```

```python
"""
Q4. Read the attached insur-2d.csv file using Pandas library. Compute the 
multivariate regression coefficient for given csv file data. Print the 
coefficients.
"""
import pandas as pd
data=pd.read_csv('/content/sample_data/insur-2d.csv')
data.insert(0,'const',1)
data.columns =['const', 'input 1', 'input 2', 'output']
input=np.matrix([data['const'].tolist(),data['input 1'].tolist(),
                 data['input 2'].tolist()]).T
output=np.matrix(data['output'].tolist()).T
beta_1=coefficient(input,output)
print('B0: ',beta_1[0],'\nB1: ',beta_1[1],'\nB2 ',beta_1[2])
```

```python
"""
Q5. Predict the three outputs for unknown three pair of inputs.
"""
x1=np.array([1,7.5,131])
o1=np.dot(x1,beta_1)
print("For input: [",x1[1],",",x1[2],"]- output is : ",float(o1[0]),"")

x2=np.array([1,3.5,45])
o2=np.dot(x2,beta_1)
print("For input: [",x2[1],",",x2[2],"]- output is : ",float(o2[0]),"")

x3=np.array([1,7.2,27])
o3=np.dot(x3,beta_1)
print("For input: [",x3[1],",",x3[2],"]- output is : ",float(o3[0]),"")
```
