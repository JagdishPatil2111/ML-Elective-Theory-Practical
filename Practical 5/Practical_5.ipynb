Aim: Write a python program to implement discriminant function and decision boundary.

Note:
In this practical, 'T' is used for transpose of a vector.

```python
"""
Q1. Classify test sample x_test = [1, 2]T using discriminant function for given 
data: m1 = [3, 6]T, m2= [3, -2]T, S1 = [[0.5, 0], [0, 2]], S2 = [[2, 0],[0, 2]], 
p(w1)=0.5, p(w2)=0.5.
"""
import numpy as np
import matplotlib.pyplot as plt

m1 = np.array([[3,6]]).T
m2 = np.array([[3,-2]]).T
s1 = np.array([[0.5,0],[0,2]])
s2 = np.array([[2,0],[0,2]])
p_w1 = 0.5
p_w2 = 0.5
x_test = np.array([[1,2]]).T

def W(s):
  return -0.5*np.linalg.inv(s)
def w(s,m):
  return np.dot(np.linalg.inv(s),m)
def wi(m,s,p_w):
  i_dot = w(s,m)
  t1 = (-0.5*np.dot(m.T, i_dot))-0.5*np.log(np.linalg.det(s))
  t2 = np.log(p_w)
  return t1+t2

g1 = np.dot(np.dot(x_test.T, W(s1)), x_test) + np.dot((w(s1,m1)).T,x_test) + wi(m1,s1,p_w1)
g2 = np.dot(np.dot(x_test.T, W(s2)), x_test) + np.dot((w(s2,m2)).T,x_test) + wi(m2,s2,p_w2)
if g1[0]>g2[0]:
  print("x_text belongs to class 1")
else:
  print("x_text belongs to class 2")
```

```python
"""
Q2. Classify test sample x_test = [2, 2]T for above data.
"""
import numpy as np
import matplotlib.pyplot as plt
m1 = np.array([[3,6]]).T
m2 = np.array([[3,-2]]).T
s1 = np.array([[0.5,0],[0,2]])
s2 = np.array([[2,0],[0,2]])
p_w1 = 0.5
p_w2 = 0.5
x_test = np.array([[2,2]]).T
def W(s):
  return -0.5*np.linalg.inv(s)
def w(s,m):
  return np.dot(np.linalg.inv(s),m)
def wi(m,s,p_w):
  i_dot = w(s,m)
  t1 = (-0.5*np.dot(m.T, i_dot))-0.5*np.log(np.linalg.det(s))
  t2 = np.log(p_w)
  return t1+t2

g1 = np.dot(np.dot(x_test.T, W(s1)), x_test)+np.dot((w(s1,m1)).T,x_test)+wi(m1,s1,p_w1)
g2 = np.dot(np.dot(x_test.T, W(s2)), x_test)+np.dot((w(s2,m2)).T,x_test)+wi(m2,s2,p_w2)
if g1[0]>g2[0]:
  print("x_text belongs to class 1")
else:
  print("x_text belongs to class 2")
```

```python
"""
Q3. Find the Decision boundary for subtask 1. Take x1=np.arange(-6,6,0.5) & 
x2=np.arange(-6,6,0.5) and use np.meshgrid for converting 1D data to 2D data. 
Plot the 3D surface of g1(x), g2(x) and g.
"""
x1=np.arange(-6,6,0.5)
x2=np.arange(-6,6,0.5)
X11, X12 = np.meshgrid(x1,x1)
X21, X22 = np.meshgrid(x2,x2)
X1 = np.array([X11,X12])
X2 = np.array([X21,X22])

def s_term(X,m,s):
  s_inv = np.linalg.inv(s)
  p1 = np.dot((X-np.expand_dims(m,axis=-1)).T, s_inv)
  p2 = np.einsum('ijk, kji->ij',p1,(X-np.expand_dims(m,axis=-1)))
  return -0.5*p2
gx1 = s_term(X1,m1,s1)+np.log(p_w1)-0.5*np.log(np.linalg.det(s1))
gx2 = s_term(X2,m2,s2)+np.log(p_w2)-0.5*np.log(np.linalg.det(s2))
g = gx1-gx2
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X21, X22, gx1, cmap='Greens')
ax.plot_surface(X21, X22, gx2, cmap='Blues')
ax.plot_surface(X21, X22, g, cmap='Reds')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('pdf')
ax.set_title('Decision Boundary')
```
