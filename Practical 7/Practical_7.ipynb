Aim: Write a python program to plot Receiver Operating Characteristics (ROC) curves.

```python
"""
Q1. Plot the “No Overlap” condition of two Gaussian distributions. Also plot the 
ROC curve for same. Give proper labeling to plot.
"""
import numpy as np
import matplotlib.pyplot as plt
u1= 1
u2= 10
s1= 1
s2= 1
x1= np.arange(-5,15,0.01)
x2= np.arange(-5,15,0.01)
p_x1= np.zeros(len(x1))
p_x2= np.zeros(len(x2))
for i in range(len(x1)):
  p_x1[i]= np.exp(-(x1[i]-u1)**2/(2*s1**2))/np.sqrt(2*np.pi*s1**2)
  p_x2[i]= np.exp(-(x2[i]-u2)**2/(2*s2**2))/np.sqrt(2*np.pi*s2**2)
plt.title('Gaussian Curve')
plt.plot(x1,p_x1,x2,p_x2)
plt.xlabel('X')
plt.ylabel('p_plt.figureX')
plt.show()
tp= np.zeros(len(x1))
fp= np.zeros(len(x2))
for i in range(len(x1)):
  tp[i]= sum(p_x2[x2>=[i]])
  fp[i]= sum(p_x1[x1>=[i]])
plt.figure
plt.title('ROC Curve No overlapping')
plt.plot(fp/100,tp/100)
plt.xlabel('fp/100')
plt.ylabel('tp/100')
plt.show()
```

```python
"""
Q2. Plot the “Partial Overlap” condition of two Gaussian distributions. Also 
plot the ROC curve for same.
"""
u5= 1
u6= 3
s5= 1
s6= 1

x5= np.arange(-10,20,0.01)
x6= np.arange(-10,20,0.01)
p_x5= np.zeros(len(x5))
p_x6= np.zeros(len(x6))
for i in range(len(x5)):
  p_x5[i]= np.exp(-(x5[i]-u5)**2/(2*s5**2))/np.sqrt(2*np.pi*s5**2)
  p_x6[i]= np.exp(-(x6[i]-u6)**2/(2*s6**2))/np.sqrt(2*np.pi*s6**2)
plt.figure
plt.title('Gaussian Curve - Partial Overlap')
plt.plot(x5,p_x5,x6,p_x6)
plt.xlabel('X')
plt.ylabel('p_plt.figureX')

plt.show()
tp1= np.zeros(len(x5))
fp1= np.zeros(len(x6))
for i in range(len(x5)):
  tp1[i]= sum(p_x6[x6>=x6[i]])
  fp1[i]= sum(p_x5[x5>=x5[i]])
plt.figure
plt.title('ROC Curve partial overlapping')
plt.plot(fp1/100,tp1/100)
plt.xlabel('fp/100')
plt.ylabel('tp/100')
plt.show()
```
